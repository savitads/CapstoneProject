```python
import pyodbc
import pandas as pd
conn=pyodbc.connect('Driver={SQL Server};'
                    'Server=*****************;'
                    'Database=Linkedin_Database;'
                    'Trusted_Connection=yes;')
query = f"select * from [Linkedin_Database].[dbo].[Jobdata]"
pdf = pd.read_sql(query, conn)
from pyspark.sql import SparkSession
 
```

```python
df=pd.DataFrame(pdf) 
df.head()
```

## Data cleaning of Job location

```python
# Convert the value of 'Name' column to lowercase
df['jobTitle'] = df['jobTitle'].str.lower()
df[['jobLocation']]
dfJob= df['jobLocation'].str.split(',', expand=True)
df['jobLocation'] = dfJob[0]
df
```

# Data cleaning on Job function and Job Industry : Removing the words after comma and filling the None to Any for Job function and Job Industry

```python
dfJobFunction= df['jobFunctions'].str.split(',', expand=True)
dfJobIndustries= df['jobIndustries'].str.split(',', expand=True)
df['jobFunctions'] = dfJobFunction[0]
df['jobIndustries'] = dfJobIndustries[0]
```

```python
df.head(5)
```

```python
df[['jobFunctions', 'jobIndustries']] = df[['jobFunctions', 'jobIndustries']].fillna('Any')
df.head(5)
```

```python
# Using DataFrame.filter() method.
df2 = df.filter(['jobTitle','jobFunctions', 'jobIndustries', 'jobLocation'], axis=1)
```

```python
df2.head()
```

```python
df2[['jobFunctions', 'jobIndustries']] = df[['jobFunctions', 'jobIndustries']].fillna('Any')
```

```python
df2.head()
```

# 4. Which job role will be having highest demand

```python
df_jobrole_demand=df2.groupby(['jobIndustries', 'jobFunctions', 'jobTitle']).size().sort_values(ascending=False)
.reset_index(name='Count') 
```

```python
cursor = conn.cursor() 
cursor.execute('''drop table if exists JobroleDemand''')
cursor.execute('''
		CREATE TABLE JobroleDemand (
			jobIndustries varchar(MAX),
			jobFunctions varchar(MAX),
            jobTitle varchar(MAX),
            Count smallint)
 
               ''');

conn.commit()
#Save the data back to database 
#Insert Dataframe into SQL Server:
for index, row in df_jobrole_demand.iterrows():
     cursor.execute("INSERT INTO JobroleDemand (jobIndustries,jobFunctions,jobTitle,Count) values(?,?,?,?)", row.jobIndustries,
                    row.jobFunctions,row.jobTitle, row.Count)
conn.commit()
cursor.close()
query = f"select * from [Linkedin_Database].[dbo].[JobroleDemand]"
pdf = pd.read_sql(query, conn)
pdf
```

# Which of the Job Industry and job function is in Demand

```python
df_job_industry_demand=df2.groupby(['jobIndustries', 'jobFunctions']).size().sort_values(ascending=False)
.reset_index(name='Count') 
```

```python
cursor = conn.cursor() 
cursor.execute('''drop table if exists JobIndustDemand''')
cursor.execute('''
		CREATE TABLE JobIndustDemand (
			jobIndustries varchar(MAX),
			jobFunctions varchar(MAX),
            Count smallint)
 
               ''');
conn.commit()
#save the data back to database
# Insert Dataframe into SQL Server:
for index, row in df_job_industry_demand.iterrows():
     cursor.execute("INSERT INTO JobIndustDemand (jobIndustries,jobFunctions,Count ) values(?,?,?)", row.jobIndustries,
                    row.jobFunctions,row.Count )
conn.commit()
cursor.close()
query = f"select * from [Linkedin_Database].[dbo].[JobIndustDemand]"
pdf = pd.read_sql(query, conn)
pdf
```

#  Finding which location has more number of jobs 

```python
df_jobLoc_count=df2.groupby(['jobLocation', 'jobIndustries']).size().sort_values(ascending=False).reset_index(name='Count') 
```

```python
cursor = conn.cursor() 
cursor.execute('''drop table if exists JobLocation_demand''')
cursor.execute('''
		CREATE TABLE JobLocation_demand (
			jobLocation varchar(MAX),
			jobIndustries varchar(MAX),
            Count smallint)
 
               ''');
conn.commit()
#save the data back to database
# Insert Dataframe into SQL Server:
for index, row in df_jobLoc_count.iterrows():
     cursor.execute("INSERT INTO JobLocation_demand (jobLocation,jobIndustries,Count ) values(?,?,?)", row.jobLocation,
                    row.jobIndustries,row.Count )
conn.commit()
cursor.close()
query = f"select * from [Linkedin_Database].[dbo].[JobLocation_demand]"
pdf = pd.read_sql(query, conn)
pdf
```
